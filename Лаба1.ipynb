```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df_salary=pd.read_csv('/content/drive/MyDrive/salary/Data Science Salary 2021 to 2023.csv')
```

```python
df=df_salary
item0 = df.shape[0]
df = df.drop_duplicates()
item1 = df.shape[0]
print(f"There are {item0-item1} duplicates found in the dataset")
```

```python
df=df_salary.drop_duplicates()
df.info()
```

```python
avg_salary_by_job_title = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False)

top_15=avg_salary_by_job_title.head(10)
plt.figure(figsize=(12,7))
top_15.plot(kind='barh', color='green', edgecolor='black')

plt.title('15 Наиболее оплачиваемых профессий в Data Science',fontsize=14)
plt.xlabel('Зарплата в USD',fontsize=14)
plt.ylabel('Название профессий',fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
```

```python
avg_salary_by_experience = df.groupby("experience_level")["salary_in_usd"].mean().sort_values(ascending=False)

avg_salary_by_experience.plot(kind='bar',color='purple', edgecolor='black')

plt.title('Средняя заработная плата в зависимости от опыта')
plt.xlabel('Опыт')
plt.ylabel('Средняя зараплата')

print('Средняя зарплата на должностях начального уровня в области науки о данных составляет', avg_salary_by_experience['EN'],'USD')
print('Средняя зарплата на должностях среднего звена в области науки о данных составляет',avg_salary_by_experience['MI'],'USD' )
print('Средняя зарплата на должностях старшего уровня в области науки о данных составляет',avg_salary_by_experience['SE'],'USD')
print('Средняя зарплата на руководящих должностях в области науки о данных составляет',avg_salary_by_experience['EX'],'USD')
```

```python
avg_salary_by_company_size=df.groupby('company_size')['salary_in_usd'].mean().sort_values(ascending=False)

avg_salary_by_company_size.plot(kind='barh',color='coral',edgecolor='black')
plt.title('Средняя зарплата по размеру компании')
plt.xlabel('Средняя зарплата в долларах США')
plt.ylabel('Размер компании')
```
